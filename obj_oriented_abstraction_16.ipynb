#abstraction

```python
from abc import ABC,abstractmethod
class car(ABC):
    @abstractmethod
    def airbag(self):
        pass
```

```python
class nexon(car):
    def airbag(self):
        return "airbag loaded sucessfully"
    def mode_of_car(self):
        return "nexon is auto mode"
```

```python
ob1=nexon()
```

```python
ob1.airbag()
```

```python
ob1.mode_of_car()
```

```python
class cricket(ABC):
    @abstractmethod
    def cricket_kit(self):
        pass
```

```python
class bat(cricket):
    def cricket_kit(self):
        return "cricket kit is must be needed to play"
    def bat(self):
        return "The branded bat must be needed"
```

```python
ob2=bat()
```

```python
ob2.bat()
```

```python
ob2.cricket_kit()
```
