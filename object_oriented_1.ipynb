#incapsulation

```python
class car():
    def __init__(self,fuel,color,engine,mode):
        self.fuel=fuel#public
        self._color=color#protected
        self.__engine=engine#private
        self.mode=mode
tata_punch=car("petrol+cng","Black",1420,"Auto")
```

```python
tata_punch.fuel
```

```python
tata_punch._color
```

```python
tata_punch._car__engine
```

#calculator

```python
class cal():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        print(self.a+self.b)
    def mul(self):
        print(self.a*self.b)
    def div(self,c):
        print((self.a/self.b)*c)
```

```python
obje=cal(20,56)
```

```python
obje.add()
```

```python
obje.div(54)
```

```python
obje.mul()
```

#inheritance

```python
class parent():
    def dancing(self):
        print("Parent with dancing skills")
```

```python
class child(parent):
    def cricket(self):
        print("child with cricketing skills")
```

```python
ob1=parent()
```

```python
ob1.dancing()
```

```python
ob2=child()
```

```python
ob2.cricket()
```

```python
class next_child(child):
    def studying(self):
        print("im child with all skills")
```

```python
nc=next_child()
```

```python
nc.cricket()
```
